```python
from pathlib import Path
path = Path('./log')
```

```python
files = [file for file in path.rglob('*.pkl')]
files
```

```python
import pickle as pkl
for file in files:
    with file.open('rb') as f:
        d = pkl.load(f)
        for key in d:
            if type(d[key]) == list:
                print(f'{key}: {len(d[key])}')
            else:
                print(f'{key}: {d[key]}')
        #print(d)
```
